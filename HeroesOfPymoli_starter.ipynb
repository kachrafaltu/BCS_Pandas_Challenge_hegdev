```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

print(purchase_data.count())
purchase_data.head(20)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
number_of_unique_items = purchase_data["Item ID"].nunique()
average_price = purchase_data["Price"].mean()
purchase_count = purchase_data["Purchase ID"].count()
total_purchases = purchase_data["Price"].sum()
purchase_summary_df = pd.DataFrame( {" Number of Unique Items":[number_of_unique_items],
                                    "Average Price": average_price,
                                    "Number of Purchases": purchase_count,
                                    "Total Revenue": total_purchases})
purchase_summary_df["Average Price"]=purchase_summary_df["Average Price"].astype(float).map("${:,.2f}".format)
purchase_summary_df["Total Revenue"]=purchase_summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
purchase_summary_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_unique_df = purchase_data[["Gender","SN"]]
purchase_unique_df.count()
purchase_unique_df=purchase_unique_df.drop_duplicates()
purchase_unique_grp = purchase_unique_df.groupby(["Gender"])
gender_count = purchase_unique_grp.count()
gender_count_df = gender_count_df.rename(columns={"SN":"Total Count"})
gender_count_sum = gender_count_df["Total Count"].sum()
gender_count_df["Percentage of Players"] = gender_count_df["Total Count"]*100/gender_count_sum
gender_count_df["Percentage of Players"]=gender_count_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)
gender_count_df.head()
```

Majority of players are male.


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_gen_grp = purchase_data.groupby(["Gender"])
purchase_gen_count_ser = purchase_gen_grp["SN"].count()
purchase_gen_avgpp_ser = purchase_gen_grp["Price"].sum()/purchase_gen_grp["SN"].count()
purchase_gen_total_ser = purchase_gen_grp["Price"].sum()
purchase_gen_count_per = purchase_gen_grp.nunique()["SN"]
purchase_gen_avg_per = purchase_gen_total_ser/purchase_gen_count_per
purchase_gen_avg_per = purchase_gen_avg_per.astype(float).map("${:,.2f}".format)
purchase_gen_total_ser = purchase_gen_total_ser.astype(float).map("${:,.2f}".format)
purchase_gen_avgpp_ser = purchase_gen_avgpp_ser.astype(float).map("${:,.2f}".format)
gender_summary_df = pd.DataFrame({"Purchase Count": purchase_gen_count_ser,
                                    "Average Purchase Price": purchase_gen_avgpp_ser,
                                    "Total Purchase Value" : purchase_gen_total_ser,
                                    "Avg Total Purchase per Person" : purchase_gen_avg_per
                                 })

gender_summary_df
```

Males seem to be the major purchasers by an overwhelming amount.

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
player_age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
bin_descriptions = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],player_age_bins, labels=bin_descriptions)
purchase_age_group = purchase_data.groupby(["Age Group"])
total_count_players = purchase_age_group["SN"].nunique()
percentage_age_group = (total_count_players/total_players) * 100
percentage_age_group = percentage_age_group.astype(float).map("{:,.2f}%".format)
age_summary_df = pd.DataFrame({"Total Count": total_count_players,"Percentage of Players": percentage_age_group})

age_summary_df
```

Most of the purchasers are in the 20-24 age group.
